```python
import pandas as pd

df_train = pd.read_csv('Training.csv')
df_train.head()
```

```python
import pandas as pd

df_test = pd.read_csv('Testing.csv')
df_test.head()
```

```python
df_train.shape
```

```python
df_test.shape
```

```python
print("Number of NaN values in train data is "+str(df_train.isnull().sum().sum()))
print("Number of NaN values in test data is "+str(df_test.isnull().sum().sum()))
```

```python
from sklearn import preprocessing
df_categorical = df_train.select_dtypes(include=['object'])
df_categorical.head()
```

```python
le = preprocessing.LabelEncoder()
df_categorical = df_categorical.apply(le.fit_transform)
df_categorical.head()
```

```python
df1 = df_train.drop(df_categorical.columns,axis=1)
df1 = pd.concat([df1,df_categorical],axis=1)
df1.head()
```

```python
from sklearn import preprocessing
df_categorical1 = df_test.select_dtypes(include=['object'])
df_categorical1.head()
```

```python
le = preprocessing.LabelEncoder()
df_categorical1 = df_categorical1.apply(le.fit_transform)
df_categorical1.head()
```

```python
df2 = df_test.drop(df_categorical1.columns,axis=1)
df2 = pd.concat([df2,df_categorical1],axis=1)
df2.head()
```

```python
x_train = df1.drop(["prognosis"], axis = 1)
y_train = df1["prognosis"]

x_test = df2.drop(["prognosis"], axis = 1)
y_test = df2["prognosis"]

x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=100)
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
